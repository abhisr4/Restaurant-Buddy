```python
import pandas as pd
import csv
import numpy as np
import json
from math import sqrt

df=pd.read_csv("KolkataData.csv")
#print(df.columns)

data=np.asarray(df[['name','lat','long']])
```

```python
data=np.asarray(df[['name','lat','long']])
#convert lat and long(str) to float to do numerical computations
dataDict={}
for item in range(len(data)):
    lati=int(float(data[item][1])*1000000)
    longi=int(float(data[item][2])*1000000)

    data[item][1]=lati
    data[item][2]=longi

    #create a dictionary to create id for a res
    dataDict[item]=data[item][0]
    data[item][0]=item
    
with open('outfile.txt','wb') as f:
    for line in data:
        np.savetxt(f, line,fmt='%0.2f')
```

```python
#TOP 100 RES IS CALCULATED BY calculation.cpp
#stored in result.txt, all the number represents the res_id(started from 0)

file=open("result.txt","r")
str=file.read()
file.close()
```

```python
nearestMatrix=[]
a=str.splitlines()
for i in a:
    x=i.split(sep=' ')
    nearestMatrix.append(x)
for resid in range(len(nearestMatrix)):
    for item in range(len(nearestMatrix[resid])):
        x=nearestMatrix[resid][item]
        if x!='':
            nearestMatrix[resid][item]=dataDict[int(x)]
```

```python
mat=[]
for i in nearestMatrix:
    string=""
    for j in i:
        string+=j
        string+=", "
    mat.append(string)
mat=pd.DataFrame(mat,columns=['nearest'])
mat
```
